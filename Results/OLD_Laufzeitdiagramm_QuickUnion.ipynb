# Laufzeit vergleiche für QuickUnion
Algorithmen und Datenstrukturen Praktikum 2 

Es werdend die Laufzeiten für die Algorithmen Quick Union und Quick Union mit Pfadkomrpession verglichen. 
Die laufzeitmessung wurde in Java durchgeführt. 
Hier findet nur die Auswertung der Messungen Stadt. 

```python
# Notwenige Bibliotheken/Werkzeuge einbinden
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import numpy as np 
```

```python
# Messdaten Einlesen
df_normal = pd.read_csv('2024-06-05_results_QuickUnion_without_compression.csv',sep=',', lineterminator='\n',index_col='Trial')
df_compression = pd.read_csv('2024-06-05_results_QuickUnion_with_compression.csv',sep=',', lineterminator='\n',index_col='Trial')


print(df_normal.head(11))
print(df_compression.head(11))
```

```python
# Diagram vorbereiten 

# Styledefinitionen für das Diagramm
# sns.set_theme(style="darkgrid")
matplotlib.rcParams['font.family'] = 'serif'
matplotlib.rcParams['font.size'] = 12


fig=plt.figure(figsize=(12,8))
ax = plt.gca()

# Plotten 
plt.plot(df_normal['N'],df_normal['Time'],label='RunTime normaler QuickUnion')
plt.plot(df_compression['N'],df_compression['Time'],label='RunTime komprimierender QuickUnion')

# Achsenskallierung 
plt.yscale("log")
plt.xscale("log")

# Legende,  Title, Achsenbeschriftungen im Diagramm anzeigen.
plt.title("Quick-Union\ngemessene Laufzeiten im Vergleich")
plt.gca().legend( loc="upper center")
plt.gca().set_xlabel("Anzahl der Elemente / Listengröße")
plt.gca().set_ylabel("Laufzeit in ns")
# plt.gca().ticklabel_format(useMathText=True)
```

```python
# Berechne Steigung und y-Schnittpunkt der Geraden für Logarithmische Werte

# Normaler QU
x_n = np.log(df_normal['N'])
y_n = np.log(df_normal['Time'])
#   Berechne Steigung und y-Schnittpunkt m,n für  y = mx + n 
slope_intercept_n = np.polyfit(x_n,y_n,1)
print('Steigung m_normal = '+str(slope_intercept_n[0])+'\nSchnittpunkt_normal n = '+str(slope_intercept_n[1]))


# Komprimierter QU
x_c = np.log(df_compression['N'])
y_c = np.log(df_compression['Time'])
#   Berechne Steigung und y-Schnittpunkt m,n für  y = mx + n 
slope_intercept_c = np.polyfit(x_c,y_c,1)
print('Steigung m_komp = '+str(slope_intercept_c[0])+'\nSchnittpunkt_komp n = '+str(slope_intercept_c[1]))

```
